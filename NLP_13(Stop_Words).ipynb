```python
f=df[ df["topics"].str.len()!=0 ]
```

```python
def preprocess(text):
    d=nlp(text)
    n=[i.text for i in d if not i.is_stop and not i.is_punct]
    return " ".join(n)
```

```python
f.contents.iloc[0]
```

```python
len(f.contents.iloc[0])
```

```python
f['contents_n']=f['contents'].apply(preprocess) 
```
